## Data Cleaning
Data cleaning means fixing bad data in your data set.

Bad data could be:

* Empty cells
* Data in wrong format
* Wrong data
* Duplicates

#### Pandas - Cleaning Empty Cells
Empty Cells
Empty cells can potentially give you a wrong result when you analyze data.

Remove Rows
One way to deal with empty cells is to remove rows that contain empty cells.

This is usually OK, since data sets can be very big, and removing a few rows will not have a big impact on the result.

```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
print(df)
```

* The DataFrames object has a method called info(), that gives you more information about the data set.

Example: Print information about the data:

```python
print(df.info())
```

```python
df1 = df.dropna()
print(df1)
```

#### Replace Empty Values
Another way of dealing with empty cells is to insert a new value instead.

This way you do not have to delete entire rows just because of some empty cells.

The fillna() method allows us to replace empty cells with a value:

Example
Replace NULL values with the number 130:


```python
import numpy as np
import pandas as pd

df = pd.DataFrame({"name": ['Alfred', pd.NA, 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
print(df)
```

```python
df.fillna(130, inplace = True)
df
```

#### Replace Only For Specified Columns
The example above replaces all empty cells in the whole Data Frame.

To only replace empty values for one column, specify the column name for the DataFrame:

Example
Replace NULL values in the "Calories" columns with the number 130:


```python
df = pd.DataFrame({"name": ['Alfred', pd.NA, 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})

df
```

```python
df["name"].fillna("Batman", inplace = True)
df
```

#### Replace Using Mean, Median, or Mode
A common way to replace empty cells, is to calculate the mean, median or mode value of the column.

Pandas uses the mean() median() and mode() methods to calculate the respective values for a specified column:

Example
Calculate the MEAN, and replace any empty values with it:



```python
df = pd.DataFrame({"item": ["milk","sugar","chips","coffee"],
                   "quantity": [2,1,5,4],
                   "price": [67, pd.NA,45,73]})

df
```

#### Pandas - Fixing Wrong Data
Wrong Data
"Wrong data" does not have to be "empty cells" or "wrong format", it can just be wrong, like if someone registered "199" instead of "1.99".

Sometimes you can spot wrong data by looking at the data set, because you have an expectation of what it should be.

If you take a look at our data set, you can see that in row 7, the duration is 450, but for all the other rows the duration is between 30 and 60.

It doesn't have to be wrong, but taking in consideration that this is the data set of someone's workout sessions, we conclude with the fact that this person did not work out in 450 minutes.

#### Replacing Values
One way to fix wrong values is to replace them with something else.

In our example, it is most likely a typo, and the value should be "45" instead of "450", and we could just insert "45" in row 7:

ExampleGet your own Python Server
Set "Duration" = 45 in row 7:



```python
df.loc[7, 'Duration'] = 45
```

```python
df
```

##### For small data sets you might be able to replace the wrong data one by one, but not for big data sets.

To replace wrong data for larger data sets you can create some rules, e.g. set some boundaries for legal values, and replace any values that are outside of the boundaries.

Example
Loop through all values in the "Duration" column.

If the value is higher than 120, set it to 120:



```python
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120
```

Another way of handling wrong data is to remove the rows that contains wrong data.

This way you do not have to find out what to replace them with, and there is a good chance you do not need them to do your analyses.

Example
Delete rows where "Duration" is higher than 120:


```python
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)
```

### Read Text file
To read a normal text file in pandas, you can use the read_csv function with appropriate parameters to handle the text file format. Even though the function is named read_csv, it can be used to read various delimited text files, not just CSV files. Here's how you can read a normal text file using pandas:

```python
# Read a text file using read_csv
df = pd.read_csv('text_data.txt', delimiter='\t')  # Replace 'file.txt' with the actual path to your text file
print(df)
```

In the code above, delimiter='\t' specifies that the file is tab-delimited. You can replace '\t' with other delimiters if your text file is delimited differently.

If your text file doesn't have any delimiters and each line represents a separate entry, you can use the following approach:

```python
# Read the text file and create a DataFrame
data = []

with open('text_data2.txt', 'r') as file:
    lines = file.readlines()
    
    for line in lines:
        fields = line.strip().split()
        data.append(fields)

df = pd.DataFrame(data,columns=['Name', 'Age', 'City'])
print(df)
```

In this example, each line represents a record, and the fields within each record are not separated by delimiters. The fields are simply consecutive values.

You can read this type of data into a pandas DataFrame by splitting each line into fields and then creating a DataFrame from the resulting list of fields.

Here,each line of the text file is split into fields using the split() method. The resulting fields are then added to the data list, which is used to create a DataFrame. The columns parameter specifies the column names for the DataFrame.


#### The Fibonacci series
is a sequence of numbers where each number is the sum of the two preceding ones, usually starting with 0 and 1.

```python
def generate_fibonacci_with_loop(n):
    fibonacci_series = [0, 1]  # Starting with the first two numbers in the sequence

    for i in range(2, n):
        next_number = fibonacci_series[i - 1] + fibonacci_series[i - 2]
        fibonacci_series.append(next_number)

    return fibonacci_series[:n]  # Return the first 'n' numbers in the Fibonacci series

# Generate and print the first 10 numbers in the Fibonacci series
n = 10
fibonacci_sequence = generate_fibonacci_with_loop(n)
print(f"Fibonacci Series (First {n} numbers) using loop method:", fibonacci_sequence)
```

#### Why we need Seaborn ?

Seaborn is a higher-level data visualization library built on top of Matplotlib. While Matplotlib provides a wide range of customization and control over plots, Seaborn offers a more streamlined and aesthetically pleasing interface for creating complex statistical visualizations. Here's why you might choose Seaborn over Matplotlib or when it is beneficial to use both:

* Simpler Syntax: Seaborn simplifies the process of creating complex plots by providing higher-level functions with sensible defaults. This reduces the amount of code needed to produce attractive visualizations.

* Built-in Themes and Styles: Seaborn comes with built-in themes and color palettes that can greatly enhance the aesthetics of your plots. It allows you to easily change the overall look of your plots with just a single line of code.

* Statistical Plotting: Seaborn is designed to work well with statistical data, making it easier to create various types of statistical visualizations such as scatter plots, box plots, violin plots, and more.

* Data Exploration: Seaborn provides functions for quickly exploring data relationships and distributions, making it convenient for initial data analysis.

* Complex Plot Types: Seaborn simplifies the creation of more complex plots, such as pair plots, joint plots, and faceted plots, which can be time-consuming to set up in Matplotlib.

* Integration with Pandas: Seaborn can work seamlessly with Pandas DataFrames, making it easy to visualize data directly from your datasets.

* Statistical Visualization: Seaborn's emphasis on statistical visualizations and its integration with statistical functions can help you uncover insights in your data more easily.

While Seaborn is a powerful choice for many visualization tasks, there are cases where Matplotlib might still be preferred. Matplotlib offers finer control over customization and is suitable for creating highly customized or unconventional plots. Moreover, Seaborn is built on top of Matplotlib, so you can use them together. You can start with Seaborn to quickly create attractive visualizations and switch to Matplotlib for fine-tuning or when creating specialized plots.
